### Libraries

```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import img_to_array
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import sklearn
```

### Load data

```python
df = pd.read_csv(r'Data_Entry_2017_v2020.csv')
df.head()
```

### Load images

```python
path = 'C:\\Users\\stan\\chest_images\\'
dir_ = os.listdir(path)
images, key = [], []
for (i, image) in enumerate(dir_[:5000]):
    im_dir = path + str(image)
    key.append(image)
    images.append(tf.keras.preprocessing.image.load_img(im_dir))
```

#### Resizing images

```python
px = 32
for i, image in enumerate(images):
    images[i] = image.resize((px, px))
```

#### Images to array

```python
img = []
for i, image in enumerate(images):
    img.append(img_to_array(image))
```

```python
data = np.array(img)
data.shape
```

### Age, Gender and View Position

```python
df2 = pd.DataFrame(key)
df2.rename(columns = {0 : 'Image Index'}, inplace = True)
df2 = df2.merge(df[["Finding Labels", "Patient ID", "Patient Age", "Patient Gender", "View Position"]], \
          left_on = "Image Index", right_on = df["Image Index"])
df2.head()
```

#### Gender

```python
df2["M"] = None
df2["F"] = None
for i, gender in enumerate(df2["Patient Gender"].values):
    if gender == "M":
        df2["M"][i] = 1
        df2["F"][i] = 0
    else:
        df2["M"][i] = 0
        df2["F"][i] = 1
df2.drop("Patient Gender", axis = 1, inplace = True)
df2.head()
```

```python
df[]
```

#### View position

```python
df2["PA"] = None
df2["AP"] = None
for i, pos in enumerate(df2["View Position"].values):
    if pos == "AP":
        df2["AP"][i] = 1
        df2["PA"][i] = 0
    else:
        df2["AP"][i] = 0
        df2["PA"][i] = 1
df2.drop("View Position", axis = 1, inplace = True)
df2.head()
```

```python
df2.dropna(inplace = True)
df2.head()
```

```python
df3 = df2
df3[["Classification", "Nonsense"]] = df2["Finding Labels"].str.split('|', 1, expand = True)
df3.drop(["Nonsense", "Finding Labels"], axis = 1, inplace = True)
df3.head()
```

```python
df3["Classification"] = df3[["Classification"]].apply(LabelEncoder().fit_transform)
df3.head()
```

### Train/Test

```python
Y = df3["Classification"]
classes = len(df3.Classification.unique())
test_size = 0.2

x = data
z = df3[["PA", "AP"]] # "Patient Age", "M", "F", "PA", "AP"
y = tf.keras.utils.to_categorical(Y)

x_train, x_test, z_train, z_test, y_train, y_test = \
sklearn.model_selection.train_test_split(x, z, y, test_size = test_size, random_state = 0)
```

### Model 1: DNN

#### Model structure

```python
X_img = tf.keras.Input(shape = (px, px, 3, ))
img1 = tf.keras.layers.Flatten()(X_img)
img2 = tf.keras.layers.Dense(1000, activation = 'relu')(img1)
d1 = tf.keras.layers.Dropout(0.4)(img2)
img3 = tf.keras.layers.Dense(500, activation = 'relu')(d1)
d2 = tf.keras.layers.Dropout(0.4)(img3)
img4 = tf.keras.layers.Dense(100, activation = 'relu')(d2)
d3 = tf.keras.layers.Dropout(0.4)(img4)

X_num = tf.keras.Input(shape = (2, ))
num1 = tf.keras.layers.Dense(1000, activation = 'relu')(X_num)
d4 = tf.keras.layers.Dropout(0.4)(num1)
num2 = tf.keras.layers.Dense(500, activation = 'relu')(d4)
d5 = tf.keras.layers.Dropout(0.4)(num2)
num3 = tf.keras.layers.Dense(100, activation = 'relu')(d5)
d6 = tf.keras.layers.Dropout(0.4)(num3)

conc = tf.keras.layers.concatenate([d3, d6])

out = tf.keras.layers.Dense(15, activation = 'softmax')(conc)
dnn = tf.keras.Model(inputs = [X_img, X_num], outputs = out)
dnn.summary()
```

#### Compile the model

```python
cb = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss', patience = 5)
solver = tf.keras.optimizers.Adam(learning_rate = 0.001)
loss = tf.keras.losses.CategoricalCrossentropy()
dnn.compile(loss = loss, optimizer = solver, metrics = ['accuracy'])
```

#### Train the model

```python
dnn_hist = dnn.fit([x_train, z_train], y_train, batch_size = 100, epochs = 10, validation_split = 0.25, callbacks = [cb])
```

```python
yhat_dnn = dnn.predict([x_test, z_test])
ytest_actual = np.argmax(a=y_test, axis = 1)
ytest_pred_dnn = np.argmax(a=yhat_dnn, axis = 1)
print('\nConfusion matrix:\n',sklearn.metrics.confusion_matrix(ytest_actual, ytest_pred_dnn) )
print('\nAccuracy score:', sklearn.metrics.accuracy_score(ytest_actual, ytest_pred_dnn))
```

### Model 2: CNN

```python
x_img = tf.keras.layers.Input(shape = (px,px,3)) 
conv1 = tf.keras.layers.Conv2D(32, (3, 3), strides=2, activation='relu', input_shape=(px, px, 3))(x_img) 
pool1 = tf.keras.layers.MaxPooling2D(pool_size=(3, 3), strides=1)(conv1) 
pool2 = tf.keras.layers.AveragePooling2D(pool_size = (3,3), strides = 1)(conv1)
pool_conc = tf.keras.layers.concatenate([pool1, pool2], axis = 1) 
conv2 = tf.keras.layers.Conv2D(128, (3,3), strides = 2, activation = 'relu',use_bias = False, kernel_initializer = 'glorot_uniform')(pool_conc) 
flatten =tf.keras.layers.Flatten()(conv2)
img1 = tf.keras.layers.Dense(1000, activation='relu', kernel_initializer='glorot_uniform')(flatten)
di1 = tf.keras.layers.Dropout(0.4)(img1)
img2 = tf.keras.layers.Dense(500, activation = 'relu')(di1)
di2 = tf.keras.layers.Dropout(0.4)(img2)
img3 = tf.keras.layers.Dense(200, activation = 'relu', kernel_initializer = 'glorot_uniform')(di2)
di3 = tf.keras.layers.Dropout(0.4)(img3)

out = tf.keras.layers.Dense(15, activation='softmax')(di3) 

cnn = tf.keras.Model(inputs = x_img, outputs = out)
cnn.summary()
```

#### Compile the model

```python
adam = tf.keras.optimizers.Adam(lr=0.0001)
loss=tf.keras.losses.CategoricalCrossentropy()
cnn.compile(loss=loss, optimizer=adam, metrics=['accuracy'])
```

#### Train the model

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5)
cnn_history = cnn.fit(x_train, y_train, batch_size=100, epochs=10, validation_split=0.2, callbacks=[callback]) 
```

#### Accuracy score

```python
yhat_cnn = cnn.predict(x_test)
ytest_actual = np.argmax(a=y_test, axis = 1)
ytest_pred_cnn = np.argmax(a=yhat, axis = 1)
print('\nConfusion matrix:\n',sklearn.metrics.confusion_matrix(ytest_actual, ytest_pred_cnn) )
print('\nAccuracy score:', sklearn.metrics.accuracy_score(ytest_actual, ytest_pred_cnn))
```

### Model 3: CNN + DNN

#### Model structure

```python
x_img = tf.keras.layers.Input(shape = (px,px,3)) 
conv1 = tf.keras.layers.Conv2D(32, (3, 3), strides=2, activation='relu', input_shape=(px, px, 3))(x_img) 

pool1 = tf.keras.layers.MaxPooling2D(pool_size=(3, 3), strides=1)(conv1) 
pool2 = tf.keras.layers.AveragePooling2D(pool_size = (3,3), strides = 1)(conv1)
pool_conc = tf.keras.layers.concatenate([pool1, pool2], axis = 1) 

conv2 = tf.keras.layers.Conv2D(128, (3,3), strides = 2, activation = 'relu',use_bias = False, kernel_initializer = 'glorot_uniform')(pool_conc) 
flatten =tf.keras.layers.Flatten()(conv2)

img1 = tf.keras.layers.Dense(1000, activation='relu', kernel_initializer='glorot_uniform')(flatten)
di1 = tf.keras.layers.Dropout(0.4)(img1)
img2 = tf.keras.layers.Dense(500, activation = 'relu')(di1)
di2 = tf.keras.layers.Dropout(0.4)(img2)
img3 = tf.keras.layers.Dense(200, activation = 'relu', kernel_initializer = 'glorot_uniform')(di2)
di3 = tf.keras.layers.Dropout(0.4)(img3)

x_num = tf.keras.layers.Input(shape = (2, ))  
num1 = tf.keras.layers.Dense(1000, activation = 'relu')(x_num)
d1 = tf.keras.layers.Dropout(0.4)(num1)
num2 = tf.keras.layers.Dense(500, activation = 'relu')(d1)
d2 = tf.keras.layers.Dropout(0.4)(num2)
num3 = tf.keras.layers.Dense(200, activation = 'relu')(d2)
d3 = tf.keras.layers.Dropout(0.4)(num3)

conc = tf.keras.layers.concatenate([di3, d3])

out = tf.keras.layers.Dense(15, activation='softmax')(conc) 


cd = tf.keras.Model(inputs = [x_img, x_num], outputs = out)
cd.summary()
```

#### Compile the model

```python
adam = tf.keras.optimizers.Adam(lr=0.001)
loss=tf.keras.losses.CategoricalCrossentropy()
cd.compile(loss=loss, optimizer=adam, metrics=['accuracy'])
```

#### Train the model

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5)
cd_history = cd.fit([x_train, z_train], y_train, batch_size=100, epochs=10, validation_split=0.2, callbacks=[callback]) 
```

#### Accuracy

```python
yhat = cd.predict([x_test, z_test])
ytest_actual = np.argmax(a=y_test, axis = 1)
ytest_pred_cd = np.argmax(a=yhat, axis = 1)
print('\nConfusion matrix:\n', sklearn.metrics.confusion_matrix(ytest_actual, ytest_pred_cd))
print('\nAccuracy score:', sklearn.metrics.accuracy_score(ytest_actual, ytest_pred_cd))
```
